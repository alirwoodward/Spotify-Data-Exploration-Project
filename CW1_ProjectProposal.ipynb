**<u>Song Metrics</u>**

* **Popularity**
    * Generated on a scale of 0-100 based on the number of all-time and recent streams [https://support.spotify.com/us/artists/article/how-we-generate-popular-tracks/]
* **Danceability**
    * "Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable." [5]
* **Energy**
    * "Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy." [5]
* **Key**
    * "The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1." [5]
* **Loudness**
    * "The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typically range between -60 and 0 db." [5]
* **Mode**
    * "Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0." [5]
* **Speechiness**
    * "Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks." [5]
* **Acousticness**
    * "A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic." [5]
* **Instrumentalness**
    * "Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0." [5]
* **Liveness**
    * "Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live." [5]
* **Valence**
    * "A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)." [5]
* **Tempo**
    * "The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration." [5]

## 2.2 Organizing the Data

### 2.2.1 Yearly Data

I want to be able to analyze data based on the year the song was released, need to use date time to extract yearly data so that it is performed accurately and easily

```python
from datetime import date

#adding a column to the dataframe to depict the year the track was released
df['Release Year'] = pd.to_datetime(df['Album Release Date'], format='ISO8601').dt.year
df['Release Year'].head()
```

```python
df['Release Year'].min()
```

This gave me pause because it is before 1960. After some thought I have concluded this must mean some songs from the 50's were still in the top charts during the 60's or at some point after

### 2.2.2 Decade Data

Creating bins and labels to be able to analyze music based on their decade

```python
bins= [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2023]
eraLabels =["1950's", "1960's", "1970's", "1980's", "1990's", "2000's", "2010's", "2020's"]
df["Era"] = pd.cut(df['Release Year'], bins=bins, labels=eraLabels)
```

```python
df.head()
```

### 2.2.3 Subset Data Frames by Year and Decade

Creating smaller data frames with only the data I want to analyze. In these cases, I will currently be looking at particular track features based on year and decade.

```python
yearly_data = df.groupby('Release Year').agg({'Loudness': 'mean', 'Track Name': 'count', 'Popularity': 'mean', 'Valence': 'mean', 'Danceability': 'mean'})
yearly_data = yearly_data.rename(columns={"Loudness": "Average Loudness", "Track Name": "Track Count", "Popularity": "Average Popularity", "Valence": "Average Valence", "Danceability": "Average Danceability"})
yearly_data.head()
```

```python
decade_data = df.groupby('Era').agg({'Loudness': 'mean', 'Track Name': 'count', 'Popularity': 'mean', 'Valence': 'mean', 'Danceability': 'mean'}, observed=True)
decade_data = decade_data.rename(columns={"Loudness": "Average Loudness", "Track Name": "Track Count", "Popularity": "Average Popularity", "Valence": "Average Valence", "Danceability": "Average Danceability"})
decade_data.head()
```
